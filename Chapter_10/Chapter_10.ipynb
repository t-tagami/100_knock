## 92. アナロジーデータへの適用
91で作成した評価データの各事例に対して，vec(2列目の単語) - vec(1列目の単語) + vec(3列目の単語)を計算し，そのベクトルと類似度が最も高い単語と，その類似度を求めよ．求めた単語と類似度は，各事例の末尾に追記せよ．このプログラムを85で作成した単語ベクトル，90で作成した単語ベクトルに対して適用せよ．

```python
#gensimだとアナロジーの計算元単語は除いてくれる(多分)
with open('work/91.txt') as f, open('work/92.txt', 'w') as g:
    for line in f:
        w1, w2, w3, w4 = line.strip().split(' ')
        if all(i in model for i in [w1, w2, w3]):
            sim_word, score = model.wv.most_similar(positive=[w2, w3], negative=[w1], topn=1)[0]
            g.write('{} {} {} {} {} {}\n'.format(w1, w2, w3, w4, sim_word, score))
```

```python
#85で作成した単語ベクトルの読み込み
import pickle
import scipy.spatial.distance as scd
from pprint import pprint
import numpy as np

with open('work/85.pickle', mode = 'rb') as f, open('work/f_t.pickle', mode='rb') as g:
    compressed_matrix = pickle.load(f)
    f_t = pickle.load(g)
    idx = {word:num for num, word in enumerate(f_t.keys())}
    
def word_to_vec_(word):
    if word in idx:
        return(compressed_matrix[idx[word]])
    
def calc_cos_similarity_(vec1, vec2):
    return(1 - scd.cosine(vec1, vec2))

def get_top1_similar_word(vec1, vec2, vec3):
    result = ['', 0]
    vec = vec2 + vec3 - vec1
    for word in idx.keys():
        word_vec = word_to_vec_(word)
        cos_sim = calc_cos_similarity_(vec, word_vec)
        if cos_sim > result[1] and not np.allclose(vec1, word_vec) and not np.allclose(vec2, word_vec) and not np.allclose(vec3, word_vec):
            result = [word, cos_sim]
    return result
```

```python
from tqdm import tqdm_notebook as tqdm
with open('work/91.txt') as f, open('work/92_.txt', 'w') as g:
    for line in tqdm(f):
        w1, w2, w3, w4 = line.strip().split(' ')
        if all(i in idx for i in [w1, w2, w3]):
            sim_word, score = get_top1_similar_word(word_to_vec_(w1), word_to_vec_(w2), word_to_vec_(w3))
            g.write('{} {} {} {} {} {}\n'.format(w1, w2, w3, w4, sim_word, score))
```

## 95. WordSimilarity-353での評価
94で作ったデータを用い，各モデルが出力する類似度のランキングと，人間の類似度判定のランキングの間のスピアマン相関係数を計算せよ．

```python
humans = []
scores_90 = []
scores_85 = []
with open('work/94_1.txt') as f:
    for line in f:
        w1, w2, human, score = line.strip().split('\t')
        humans.append(human)
        scores_90.append(score)
with open('work/94_2.txt') as f:
    for line in f:
        w1, w2, human, score = line.strip().split('\t')
        scores_85.append(score)
```

## 96. 国名に関するベクトルの抽出
word2vecの学習結果から，国名に関するベクトルのみを抜き出せ．

```python
import pycountry
country_list = [country.name.replace(' ', '_') for country in pycountry.countries]
country_vecs = {country:model.wv.__getitem__(country) for country in country_list 
                     if model.wv.__contains__(country)}
```

## 97. k-meansクラスタリング
96の単語ベクトルに対して，k-meansクラスタリングをクラスタ数k=5として実行せよ．

```python
from sklearn.cluster import KMeans

km_model = KMeans(n_clusters=5, random_state=114514)
km_model.fit(list(country_vecs.values()))
with open('work/97.txt', 'w') as f:
    for country, label in sorted(zip(country_vecs, km_model.labels_), key=lambda x:x[1]):
        f.write('{}\t{}\n'.format(country, label))
```

```python
from sklearn.decomposition import TruncatedSVD
compressed_vec = TruncatedSVD(2).fit_transform(list(country_vecs.values()))
cols = ['r', 'g', 'b', 'c', 'm']
plt.figure(figsize=(32,18))
for num, label, name in zip(range(len(country_vecs)), km_model.labels_, country_vecs.keys()):
    plt.plot(compressed_vec[num, 0], compressed_vec[num, 1], 'o', color = cols[label], markersize=12)
    plt.annotate(name, (compressed_vec[num, 0], compressed_vec[num, 1]), size=15)
plt.savefig('work/97.png')
```

## 98. Ward法によるクラスタリング
96の単語ベクトルに対して，Ward法による階層型クラスタリングを実行せよ．さらに，クラスタリング結果をデンドログラムとして可視化せよ．

```python
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster

ward = linkage(list(country_vecs.values()), method = 'ward')
plt.figure(figsize=(8, 64))
plt.xlabel('distance', fontsize = 16)
dendrogram(ward, orientation = 'right', labels = list(country_vecs.keys()), leaf_font_size = 16)
plt.savefig('work/98.png')
```
